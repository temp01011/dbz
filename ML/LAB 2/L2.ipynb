```python
x = x.values  # to convert 2d - 1d it is attribute (value)
```

```python
x
```

```python
x = x.reshape(117,1)
```

```python
x.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression   #class 
```

```python
regressor = LinearRegression() #object 
```

```python
regressor
```

```python
regressor.fit(x,y)   # model is train here 
```

```python
regressor.coef_  #slope value  (y=mx+c )  this is value for m #model parameter 
```

```python
regressor.intercept_ # this is the c value from slope 
```

```python
regressor.predict([[2025]])
```

```python
predicted = regressor.predict(x)  #this is the predicted by the model
```

```python
predicted
```

```python
y  #actual value
```

```python
import numpy as np
```

```python
# MEAN ABSOLUTE EROOR
np.mean (abs(y - predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,predicted)
```

```python
# MEAN SQUARED ERROR 
np.mean((y - predicted) **2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,predicted)
```

```python
# R2 SCORE
regressor.score(x,y)
```

```python
from sklearn.metrics import r2_score      #linearity 
r2_score(y, predicted) 
```

```python
# visualize model
plt.figure(figsize=(8,5))
plt.title('Temperatures YEAR and ANNUAL')
plt.xlabel('YEAR')
plt.ylabel('ANNUAL TEMP')
plt.scatter(x,y, label = "actual" , color ="r" , marker='.')
plt.plot(x, predicted, label = 'predicted' , color = "g")
plt.legend()
```
